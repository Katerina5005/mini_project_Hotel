```python
booking['total_kids'] = booking['children'] + booking['babies']
```

```python
booking.groupby('hotel')\
        .agg({'total_kids':'mean'})\
        .round(2)\
        .max()
```

```python
booking.total_kids
```

```python
booking['has_kids'] = booking.total_kids > 0
```

```python
no_kids_churn = booking.query('is_canceled == 1 and has_kids == False').shape[0]/booking.query('has_kids == False').shape[0]
no_kids_churn = round(no_kids_churn*100, 2)
```

```python
no_kids_churn
```

```python
yes_kids_churn = booking.query('is_canceled == 1 and has_kids == True').shape[0]/booking.query('has_kids == True').shape[0]
yes_kids_churn = round(yes_kids_churn*100, 2)
```
